## Importing and exporting data

See the <u>IO Tools</u> section.

### CSV

<u>Writing to a csv file:</u> using <u>`DataFrame.to_csv()`</u>

```python
df = pd.DataFrame(np.random.randint(0, 5, (10, 5)))
```

```python
df.to_csv("foo.csv")
```

```python
pd.read_csv("foo.csv")
```

### Parquet

Writing to a Parquet file:

```python
df.to_parquet("foo.parquet")
```

Reading from a Parquet file store using <u>`read_parquet()`</u>

```python
df.to_excel("foo.xlsx", sheet_name="Dataset")
```

Read from an excel file using <u>`read_excel()`</u>

```python
pd.read_excel("foo.xlsx", "Dataset", index_col=None, na_values=["NA"])
```
