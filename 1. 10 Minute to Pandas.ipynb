## Basic data structure in pandas

Pandas provides two types of classes for handling data:

1. <u>`Series`</u>: __a one-dimensional labeled array holding data of any type__
such as integers, strings, Python objects etc.

2. <u>`DataFrame`</u>: a two-dimensional data structure that holds data like a two-dimension array or a table with rows and columns.

### Getitem (<code>[]</code>)

For a <u>`DataFrame`</u>, passing a single label selects a columns and yields a <u>`Series`</u> equivalent to <u>`df.A`</u>:

```python
df["A"]
```

For a <u>`DataFrame`</u>, passing a slice `:` selects matching rows:

```python
df[0:3]
```

### Selection by label

See more in <u>Selection by Label</u> using <u>`DataFrame.loc()`</u> or <u>`DataFrame.at()`</u>.

Selecting a row matching a label:

```python
df.loc[dates[0]]
```

Selecting all rows (`:`) with a select column labels:

```python
df.loc[:,["A", "B"]]
```

For label slicing, both endpoints are included:

```python
df.loc["20241110":"20241121",["A", "B"]]
```

Selecting a single row and column label returns a scalar:

```python
df.loc[dates[0], "A"]
```

For getting fast access to a scalar (equivalent to the prior method)

```python
df.at[dates[0], "A"]
```

### Selection by Position

See more in <u>Selection by Position</u> using <u>`DataFrame.iloc()`</u> or <u>`DataFrame.iat()`</u>.

Select via the position of the passed integers

```python
df.iloc[3]
```

integer slices acts similar to Numpy/Python

```python
df.iloc[3:5, 0:2]
```

Lists of integer position locations:

```python
df.iloc[[1, 2, 4], [0, 2]]
```

For slicing rows explicitly:

```python
df.iloc[1:3, :]
```

For slicing columns explicitly:

```python
df.iloc[:, 1:3]
```

For getting a value explicitly:

```python
df.iloc[1, 1]
```

For getting dast access to a scalar (equivalent to the prior method):

```python
df.iat[1, 1]
```

## Missing Data

For NumPy data types, <u>`np.nan`</u> represents missing data. It is by default not included in computations. See the <u>Missing Data selection</u>.

Reindexing allows you to change/add/delete the index on a specified axis. This returns a copy of the data:

```python
df2 = df.reindex(index=dates[0:4], columns=list(df.columns) + ["E"])
```

```python
df2.loc[dates[0] : dates[1], "E"] = 1
df2.loc[dates[0] : dates[1], "F"] = 1
```

```python
df2
```

<u>`DataFrame.dropna()`</u> drops any rows that hava missing data:

```python
df2.dropna(how="any")
```

<u>`DataFrame.fillsna()`</u> fills missing data:

```python
df2.fillna(value=5)
```

<u>`isna()`</u> gets the boolean mask where values are <u>`nan`</u>

```python
pd.isna(df2)
```

## Grouping

By "group by" we are referring to a process involving one or more of the following steps:

* __Splitting__ the data into groups based on some criteria
* __Applying__ a function to each group independently
* __Combining__ the results into a data structure

```python
df = pd.DataFrame(
    {
        "A": ["foo", "bar", "foo", "bar", "foo", "bar", "foo", "foo"],
        "B": ["one", "one", "two", "three", "two", "two", "one", "three"],
        "C": np.random.randn(8),
        "D": np.random.randn(8),
    }
)
```

```python
df
```

Grouping by a column label, selecting column labels, and hen applying the <u>`DataFrameGroupBy.sum()`</u> function to the resulting groups:

```python
df.groupby("A")[["C", "D"]].sum()
```

Grouping by multiple columns label forms <u>`MultiIndex`</u>.

```python
df.groupby(["A", "B"]).sum()
```

## Time series

Pandas has simple, powerful, and efficient functionality for performing resampling operations during frequency conversion (e.g., converting secondly into 5-minutely data). This is extremely common in, but not limited to, financial application. See the <u>Time Series section</u>.

```python
rng = pd.date_range("1/1/2024", periods=1000, freq="s")
```

```python
ts = pd.Series(np.random.randint(0, 500, len(rng)), index=rng)
```

```python
ts.resample("5Min").sum()
```

<u>`Series.tz_localize()`</u> localizes a time series to a time zone:

```python
rng = pd.date_range("1/1/2024 00:00", periods=10, freq="D")
```

```python
ts = pd.Series(np.random.randn(len(rng)), rng)
```

```python
ts
```

```python
ts_utc = ts.tz_localize("UTC")

ts_utc
```

<u>`Series.tz_convert()`</u> converts a timezones aware time series to another time zone:

```python
ts_utc.tz_convert("US/Eastern")
```

Adding on non-fixed duration (<u>`BusinessDay`</u>) to a time series:

```python
rng
```

```python
rng + pd.offsets.BusinessDay(5)
```

## Categoricals

Pandas can include categorical data in a <u>`DataFrame`</u>. For full docs, see the <u>categorical introduction</u> and the <u>API documentation</u>.

```python
df = pd.DataFrame(
    {
        "id": [1, 2, 3, 4, 5, 6],
        "raw_grade": ["a", "b", "b", "a", "a", "e"]
    }
)
```

Converting the raw grades to a categorical data type:

```python
df["grade"] = df["raw_grade"].astype("category")
```

```python
df["grade"]
```

Rename the categories to more meaningful names:

```python
new_categories = ["very good", "good", "very bad"]
```

```python
df["grade"] = df["grade"].cat.rename_categories(new_categories)
```

Reorder the categories and simulataneously add the missing category (methods under <u>`Series.cat()`</u> return a new <u>`Series`</u> by default):

```python
df["grade"] = df["grade"].cat.set_categories(
    [
        "very bad", "bad", "medium", "good", "very good"
    ]
)
```

```python
df["grade"]
```

sorting is per order in the categories, not lexical order:

```python
df.sort_values(by="grade")
```

Grouping by a categorical column wirh <u>`observed=False`</u> also shows empty categories:

```python
df.groupby("grade", observed=False).size()
```
